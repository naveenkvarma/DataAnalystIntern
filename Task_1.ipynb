# **Importing the dataset**

```python
import pandas as pd
# Load the dataset
df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head(5)
```

# **Handling Missing Values**

```python
# Check for missing values
missing_values = df.isnull().sum()

df['Income'] = df['Income'].fillna(df['Income'].mean())
df.head(10)
```

# **Observation:**
- Missing income values were present and successfully filled using the mean strategy.

# **Removing Duplicate Records**

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
df.head(10)
```

# **Observation:**
- Several duplicate entries were identified and removed to avoid redundancy.

# **Standardizing Text Formats**

```python
# Standardize 'Education' entries
df['Education'] = df['Education'].replace({
    '2n Cycle': 'Master',
    'Basic': 'Undergraduate'
})

# Standardize 'Marital_Status' entries
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single',
    'Divorced': 'Separated',
    'Widow': 'Separated'
})
```

# **Observation:**
- Inconsistent categorical values (like 'YOLO' in Marital_Status) were standardized for analysis.

# **Converting and Parsing Dates**

```python
# Convert to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

# Extract year and month
df['Year_Customer'] = df['Dt_Customer'].dt.year
df['Month_Customer'] = df['Dt_Customer'].dt.month
df.head(10)
```

# **Observation:**
- Date fields were originally in string format and were converted to datetime objects for consistency.

# **Renaming Columns**

```python
# Rename columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(10)
```

# **Observation:**
- Column names had spaces and varied casing, which were unified to lowercase_with_underscores.

# **Checking and Fixing Data Types**

```python
# Convert 'year_birth' to integer
df['year_birth'] = df['year_birth'].astype(int)

# Calculate 'age'
df['age'] = 2025 - df['year_birth']
df.head(10)
```

# **Observation:**
- Column names had spaces and varied casing, whData types such as year_birth were verified and corrected, and an age column was derived.

# **Exporting the Cleaned Dataset**

```python
# Save the cleaned dataset
df.to_csv('cleaned_customer_personality.csv', index=False)
```
